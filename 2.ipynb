```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data.csv')
df
```

This code shows these columns have missing values

```python
for column in df.columns:
    if df[column].isna().sum() > 0:
        print(f"{column} has {df[column].isna().sum()} missing values")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[['short_name','value_eur']].sort_values(by="value_eur", ascending=False).head(40)
```

```python
groups = df[['short_name','value_eur']].sort_values(by="value_eur", ascending=False)
groups
```

```python
collection = [
    groups[groups['value_eur'] > 80000000].count(),
    groups[(groups['value_eur'] > 60000000) & (groups['value_eur'] <= 80000000)].count(),
    groups[(groups['value_eur'] > 40000000) & (groups['value_eur'] <= 60000000)].count(),
    groups[(groups['value_eur'] > 20000000) & (groups['value_eur'] <= 40000000)].count(),
    groups[(groups['value_eur'] > 10000000) & (groups['value_eur'] <= 20000000)].count()
    
]
labels = ['> 80M', '60M - 80M', '40M - 60M', '20M - 40M', '10M < 20M']
sizes = [element["value_eur"] for element in collection]

for element in collection:
    print(element)
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Player Values whose Value > 10M')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
df1 = df[['short_name','age','nationality']]
df1
```

Analysing the minimum and maximum age of players grouped by nationality

```python
age_stats = df1.groupby(by='nationality')['age'].agg(['mean','min','max']).reset_index()
age_stats
```

```python
plt.figure(figsize=(30, 10))
bar_width = 0.4
nationalities = age_stats['nationality']

plt.bar(range(len(nationalities)), age_stats['min'], width=bar_width, label='Min Age', alpha=0.6)
plt.bar([i + bar_width for i in range(len(nationalities))], age_stats['max'], width=bar_width, label='Max Age', alpha=0.6)

plt.xlabel('Nationality')
plt.ylabel('Age')
plt.title('Minimum and Maximum Age by Nationality')
plt.xticks([i + bar_width / 2 for i in range(len(nationalities))], nationalities, rotation=90)
plt.legend()
plt.grid(axis='y')

plt.show()
```

```python
df2 = df[['sofifa_id','height_cm','weight_kg','overall']]
```

```python
df2['bmi'] = df2['weight_kg'] / (df2['height_cm'] / 100) ** 2
df2
```

```python
df2[['bmi','overall']].describe()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df2, x='bmi', y='overall', alpha=0.6)

plt.title('Relationship between BMI and Overall')
plt.xlabel('BMI')
plt.ylabel('Overall')
plt.grid(True)
plt.show()
```
