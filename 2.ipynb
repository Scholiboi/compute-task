```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')
df
```

```python
df.isna().sum()
```

This code shows these columns have missing values

```python
for column in df.columns:
    if df[column].isna().sum() > 0:
        print(f"{column} has {df[column].isna().sum()} missing values")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[['short_name','value_eur']].sort_values(by="value_eur", ascending=False).head(40)
```

```python
groups = df[['short_name','value_eur']].sort_values(by="value_eur", ascending=False)
groups
```

```python
collection = [
    groups[groups['value_eur'] > 80000000].count(),
    groups[(groups['value_eur'] > 60000000) & (groups['value_eur'] <= 80000000)].count(),
    groups[(groups['value_eur'] > 40000000) & (groups['value_eur'] <= 60000000)].count(),
    groups[(groups['value_eur'] > 20000000) & (groups['value_eur'] <= 40000000)].count(),
    groups[(groups['value_eur'] > 10000000) & (groups['value_eur'] <= 20000000)].count()
    
]
labels = ['> 80M', '60M - 80M', '40M - 60M', '20M - 40M', '10M < 20M']
sizes = [element["value_eur"] for element in collection]

for element in collection:
    print(element)
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Player Values whose Value > 10M')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
